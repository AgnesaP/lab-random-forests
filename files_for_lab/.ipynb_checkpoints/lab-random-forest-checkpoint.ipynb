```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='white', context='notebook', palette='deep')
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split


import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
target = pd.read_csv("target.csv")
numerical = pd.read_csv("numerical.csv")
categorical = pd.read_csv("categorical.csv")
```

```python
data = pd.concat([numerical, categorical, target], axis = 1)
data
```

```python
data.isna().sum()
```

### The columns that are numerical

```python
numericals = data.select_dtypes(np.number)
numericals
```

### The columns that are categorical

```python
categoricals = data.select_dtypes(np.object)
categoricals
```

```python
data['TARGET_B'].value_counts()
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B'], axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = encoder = OneHotEncoder(drop='first').fit(categoricals)

encoded_categoricals = encoder.transform(categoricals).toarray()
encoded_categoricals = pd.DataFrame(encoded_categoricals)
```

```python
X = pd.concat([numericals, encoded_categoricals], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X
```

```python
y
```

### Checking for the imbalance of the data

```python
y.value_counts().to_frame().T
```

We can see that we are dealing with imbalanced data. So I will use SMOTE technique to balance the data

```python
from imblearn.over_sampling import SMOTE
x_train_resampled, y_train_resampled = SMOTE().fit_resample(X_train, y_train)
```

```python
y_train_resampled.value_counts().to_frame().T
```

Now as we can see after applying SMOTE the data are balanced.

#### Creating a Random Forest Classifier

```python

classifier = RandomForestClassifier(random_state = 42)
classifier.fit(X_train, y_train)
```

```python
print("Train Score : {}".format(classifier.score(X_train, y_train)))
print('Test Score : {}'.format(classifier.score(X_test, y_test)))
```

```python
y_pred = classifier.predict(X_test)
```

```python

accuracy_train = classifier.score(X_train, y_train)
print(f"Accuracy on the training set: {accuracy_train:.0%}")
```

```python
accuracy_test = classifier.score(X_test, y_test)
print(f"Accuracy on the test set: {accuracy_test:.0%}")
```

```python

matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
fig, ax = plt.subplots(constrained_layout=False)
plot_confusion_matrix(classifier, X_test, y_test, ax=ax)
ax.grid(b=False);
```

```python
print(classification_report(y_test, y_pred))
```

#### Creating a random forest classifier after SMOTE 

```python
classifier = RandomForestClassifier(random_state = 42)
classifier.fit(x_train_resampled, y_train_resampled )
```

```python
print("Train Score : {}".format(classifier.score(x_train_resampled, y_train_resampled)))
```

```python
accuracy_train = classifier.score(x_train_resampled, y_train_resampled)
print(f"Accuracy on the training set after SMOTE: {accuracy_train:.0%}")
```
